```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://exame.abril.com.br/noticias-sobre/desastres-naturais/")
```

```python
response.status_code
```

```python
response.text
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
soup.find('title')
```

```python

link = soup.find_all(attrs = {'class':'list-item-title'})
```

```python
Count = 1
for Linha in link:
    print (Count,"[",Linha.getText(),"]","\n")
    Count += 1
```
